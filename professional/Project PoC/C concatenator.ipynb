```python
import pandas as pd
import numpy as np
import datetime as dt

elements= pd.read_excel(r'C:\Users\fnafisa\WORKSPACE\professional\Project PoC\Input\data\Construction materials\segregated\type names.xlsx')
elements.columns= ['element_name']
elements= elements.values

path_const= r'C:\Users\fnafisa\WORKSPACE\professional\Project PoC\output\construction materials price index prediction\{}.xlsx'
path_const_output= r'C:\Users\fnafisa\WORKSPACE\professional\Project PoC\output\{}.xlsx'

res= pd.DataFrame()
```

```python
def prepare_single_const_element(filename):
    df_const= pd.read_excel(path_const.format(filename))
    df_const.columns= ['date',	'price',	'predicted_training',	'predicted_forecast',	'element_name']
    df_const= df_const.sort_values(by=['date'], ascending=True)
    df_const= df_const.loc[:,['date','price','predicted_forecast','element_name']]

    df_const_actu= df_const[ df_const['predicted_forecast'].isnull()]
    df_const_pred= df_const[~ df_const['predicted_forecast'].isnull()]

    df_const_actu['data_point']= 'actual'
    df_const_pred['data_point']= 'prediction'

    df_const_actu= df_const_actu.loc[:,['date','price','element_name','data_point']]
    df_const_pred= df_const_pred.loc[:,['date','predicted_forecast','element_name','data_point']]

    df_const_pred.columns= ['date',	'price',	'element_name', 'data_point']


    return df_const_actu.append(df_const_pred)


    
```

```python
dict_const= {}
for e in elements:
    e_name= e[0]
    dict_const[e_name]= prepare_single_const_element(e_name)

for v in dict_const:
    df_temp=dict_const[v]
    res= res.append(df_temp)

```

```python
pass_to_export= (len(res[res.data_point == 'prediction']) == 37 * 6)
print(pass_to_export)
```

```python
if(pass_to_export):
    res.to_excel(path_const_output.format('actual and prediction data'), index=False)
```
