# Active and Expired

```python
#                       ACTIVE SET
active= pd.read_excel(input_active_filepath)
# A)delete the first empty row:
#     1- transpose the dataframe
#     2- store the first column (row in the pre-transposed dataframe) and store in a variable (cols)
#     3- transpose the dataframe back to its original format
#     4- the cols to the dataframe to be its columns set and rearange its index 
active= active.T
cols= active.pop(0)
active= active.T
active.columns, active.index= cols, np.arange(len(active))

#                       EXPIRED SET
expired= pd.read_excel(input_expired_filepath)
expired= expired.T
cols= expired.pop(0)
expired= expired.T
expired.columns, expired.index= cols, np.arange(len(expired))



# B)assign status
#                       ACTIVE SET
active['status']= 'Active'
#                       EXPIRED SET
expired['status']= 'expird'



# C)create a temporary column of two values (T and S) to help in removing the "Free" memberships
#                       ACTIVE SET
ts_list= []
def t_and_s_categorization(total_emp):
    if total_emp <= 5:
        ts_list.append('T')
    else:
        ts_list.append('S')
active['#Total Employees'].apply(t_and_s_categorization)
active['TEMP']= ts_list
#                       EXPIRED SET
ts_list= []
def t_and_s_categorization(total_emp):
    if total_emp <= 5:
        ts_list.append('T')
    else:
        ts_list.append('S')
expired['Total# of Employees'].apply(t_and_s_categorization)
expired['TEMP']= ts_list



# # D)create a new column that has all the updated memberships (no Free memberships in it) and assign
#                       ACTIVE SET
new_mem_lvl= []
for i in range(len(active)):
    if 'Free' in (active.loc[i,['Membership Level']]).values[0]:
        if (active.loc[i,['TEMP']]).values[0] == 'T':
            new_mem_lvl.append((active.loc[i,['Membership Level']]).values[0].replace('Free','0-5'))
        elif (active.loc[i,['TEMP']]).values[0] == 'S':
            new_mem_lvl.append((active.loc[i,['Membership Level']]).values[0].replace('Free','6-49'))
    else:
        new_mem_lvl.append((active.loc[i,['Membership Level']]).values[0])
active['new Membership Level']= pd.Series(new_mem_lvl)
active['Membership Level']= active.pop('new Membership Level')
#                       EXPIRED SET
new_mem_lvl= []
for i in range(len(expired)):
    if 'Free' in (expired.loc[i,['Membership Level']]).values[0]:
        if (expired.loc[i,['TEMP']]).values[0] == 'T':
            new_mem_lvl.append((expired.loc[i,['Membership Level']]).values[0].replace('Free','0-5'))
        elif (expired.loc[i,['TEMP']]).values[0] == 'S':
            new_mem_lvl.append((expired.loc[i,['Membership Level']]).values[0].replace('Free','6-49'))
    else:
        new_mem_lvl.append((expired.loc[i,['Membership Level']]).values[0])
expired['new Membership Level']= pd.Series(new_mem_lvl)
expired['Membership Level']= expired.pop('new Membership Level')



# E)create order column based on the membership
#                       ACTIVE SET
order_list= []
for i in range(len(active)):
    if '0-5' in (active.loc[i,['Membership Level']]).values[0]:
        order_list.append(1)
    elif '6-49' in (active.loc[i,['Membership Level']]).values[0]:
        order_list.append(2)
    elif '50-249' in (active.loc[i,['Membership Level']]).values[0]:
        order_list.append(3)
    elif '250UP' in (active.loc[i,['Membership Level']]).values[0]:
        order_list.append(4)
    elif 'NSC' in (active.loc[i,['Membership Level']]).values[0]:
        order_list.append(5)
    elif 'ORGNZ' in (active.loc[i,['Membership Level']]).values[0]:
        order_list.append(6)
    elif 'INDVL' in (active.loc[i,['Membership Level']]).values[0]:
        order_list.append(7)
    else:
        order_list.append(999)
active['order']= pd.Series(order_list)
#                       EXPIRED SET
order_list= []
for i in range(len(expired)):
    if '0-5' in (expired.loc[i,['Membership Level']]).values[0]:
        order_list.append(1)
    elif '6-49' in (expired.loc[i,['Membership Level']]).values[0]:
        order_list.append(2)
    elif '50-249' in (expired.loc[i,['Membership Level']]).values[0]:
        order_list.append(3)
    elif '250UP' in (expired.loc[i,['Membership Level']]).values[0]:
        order_list.append(4)
    elif 'NSC' in (expired.loc[i,['Membership Level']]).values[0]:
        order_list.append(5)
    elif 'ORGNZ' in (expired.loc[i,['Membership Level']]).values[0]:
        order_list.append(6)
    elif 'INDVL' in (expired.loc[i,['Membership Level']]).values[0]:
        order_list.append(7)
    else:
        order_list.append(999)
expired['order']= pd.Series(order_list)

active_expired_length= len(active) + len(expired) # save the summation of length of the active and expired datasets

# F)Merge active and expired datasets
df_mem_cols= ['User Id', 'User Name', 'User Email', 'User Mobile', 'Auth Person Name',
       'Company Name Ar', 'Company Name En', 'Company CR', 'Membership Id',
       'Membership Level', 'Amount', 'Payment Type', 'Renewal', 'Payment Ref#',
       'Join Date', 'M Start Date', 'Expiry Date', 'Reconciliation Date',
       'Reconciliation Status', 'Payment Status', 'Account Status',
       'Classification Status', '#Saudi', '#Non-Saudi', '#Total Employees',
       'Direct Request', 'Responsible', 'Request Date', 'status', 'Fee Amount',
       'Saudi', 'Non Saudi', 'Total# of Employees', 'Request Status', 'City',
       'City Ar', 'Region', 'Region Ar', 'Order']
amount= np.concatenate((active['Amount'].values, np.full(((active_expired_length)-len(active['Amount']), 1), np.nan).flatten()))
payment_type= np.concatenate((active['Payment Type'].values, np.full(((active_expired_length)-len(active['Payment Type']), 1), np.nan).flatten()))
renewal= np.concatenate((active['Renewal'].values, np.full(((active_expired_length)-len(active['Renewal']), 1), np.nan).flatten())) 
payment_ref= np.concatenate((active['Payment Ref#'].values, np.full(((active_expired_length)-len(active['Payment Ref#']), 1), np.nan).flatten())) 
join_date= np.concatenate((active['Join Date'].values, np.full(((active_expired_length)-len(active['Join Date']), 1), np.nan).flatten())) 
m_start_date= np.concatenate((active['M Start Date'].values, np.full(((active_expired_length)-len(active['M Start Date']), 1), np.nan).flatten())) 
exp_date= np.concatenate((active['Expiry Date'].values, np.full(((active_expired_length)-len(active['Expiry Date']), 1), np.nan).flatten())) 
rec_date= np.concatenate((active['Reconciliation Date'].values, np.full(((active_expired_length)-len(active['Reconciliation Date']), 1), np.nan).flatten()))
rec_status= np.concatenate((active['Reconciliation Status'].values, np.full(((active_expired_length)-len(active['Reconciliation Status']), 1), np.nan).flatten())) 
payment_status= np.concatenate((active['Payment Status'].values, np.full(((active_expired_length)-len(active['Payment Status']), 1), np.nan).flatten()))
account_status= np.concatenate((active['Account Status'].values, np.full(((active_expired_length)-len(active['Account Status']), 1), np.nan).flatten()))
class_status= np.concatenate((active['Classification Status'].values, np.full(((active_expired_length)-len(active['Classification Status']), 1), np.nan).flatten()))


dirt_req= np.concatenate((active['Direct Request'].values, np.full(((active_expired_length)-len(active['Direct Request']), 1), np.nan).flatten()))
responsible= np.concatenate((active['Responsible'].values, np.full(((active_expired_length)-len(active['Responsible']), 1), np.nan).flatten()))
req_date= np.concatenate((active['Request Date'].values, np.full(((active_expired_length)-len(active['Request Date']), 1), np.nan).flatten()))

combiner={
    # common columns -> 10 cols | !needs to be flatten
    'User Id': pd.concat([active['User Id'], expired['User Id']], ignore_index=True).to_frame('User Id').values.flatten(),
    'User Name': pd.concat([active['User Name'], expired['User Name']], ignore_index=True).to_frame('User Name').values.flatten(),
    'User Email': pd.concat([active['User Email'], expired['User Email']], ignore_index=True).to_frame('User Email').values.flatten(),
    'User Mobile': pd.concat([active['User Mobile'], expired['User Mobile']], ignore_index=True).to_frame('User Mobile').values.flatten(),
    'Auth Person Name': pd.concat([active['Auth Person Name'], expired['Auth Person Name']], ignore_index=True).to_frame('Auth Person Name').values.flatten(),
    'Company Name Ar': pd.concat([active['Company Name Ar'], expired['Company Name Ar']], ignore_index=True).to_frame('Company Name Ar').values.flatten(),
    'Company Name En': pd.concat([active['Company Name En'], expired['Company Name En']], ignore_index=True).to_frame('Company Name En').values.flatten(),
    'Company CR': pd.concat([active['Company CR'], expired['Company CR']], ignore_index=True).to_frame('Company CR').values.flatten(),
    'Membership Id': pd.concat([active['Membership Id'], expired['Membership Id']], ignore_index=True).to_frame('Membership Id').values.flatten(),
    'Membership Level': pd.concat([active['Membership Level'], expired['Membership Level']], ignore_index=True).to_frame('Membership Level').values.flatten(),

    # only in active -> 12
    'Amount': amount,
    'Payment Type': payment_type,
    'Renewal':renewal,
    'Payment Ref#':payment_ref,
    'Join Date': join_date,
    'M Start Date': m_start_date,
    'Expiry Date': exp_date,
    'Reconciliation Date':rec_date,
    'Reconciliation Status': rec_status,
    'Payment Status': payment_status,
    'Account Status': account_status,
    'Classification Status': class_status,

    # emp tots X common columns -> 3
    '#Saudi': pd.concat([active['#Saudi'], expired['Saudi']], ignore_index=True).to_frame('#Saudi').values.flatten(),
    '#Non-Saudi': pd.concat([active['#Non-Saudi'], expired['Non Saudi']], ignore_index=True).to_frame('#Non-Saudi').values.flatten(),
    '#Total Employees': pd.concat([active['#Total Employees'], expired['Total# of Employees']], ignore_index=True).to_frame('#Total Employees').values.flatten(),

    # only in active -> 3
    'Direct Request': dirt_req,
    'Responsible': responsible,
    'Request Date': req_date,

    # new columns -> 11
    'status': pd.concat([active['status'], expired['status']], ignore_index=True).to_frame('status').values.flatten(),
    'Fee Amount': np.full((len(active)+len(expired), 1), np.nan).flatten(),
    'Saudi': np.full((len(active)+len(expired), 1), np.nan).flatten(), # Saudi
    'Non Saudi': np.full((len(active)+len(expired), 1), np.nan).flatten(), # Non Saudi
    'Total# of Employees': np.full((len(active)+len(expired), 1), np.nan).flatten(), # Total# of Employees   
    'Request Status': np.full((len(active)+len(expired), 1), np.nan).flatten(), # Request Status
    'City': np.full((len(active)+len(expired), 1), np.nan).flatten(), # City
    'City Ar': np.full((len(active)+len(expired), 1), np.nan).flatten(), # City Ar
    'Region': np.full((len(active)+len(expired), 1), np.nan).flatten(), # Region
    'Region Ar': np.full((len(active)+len(expired), 1), np.nan).flatten(), # Region Ar
    'Order': pd.concat([active['order'], expired['order']], ignore_index=True).to_frame('order').values.flatten()
}
df_merged= pd.DataFrame(combiner)

if active_expired_length == len(df_merged):
    print('Active and Expired datasets are merged :)')

# G)delete duplicate
mask= ~df_merged.duplicated(subset=['User Id'], keep=False) + (df_merged['status'] == 'Active')
df_no_duplicate= df_merged[mask]

df_no_duplicate.to_excel(output_memberships_filepath, index= False)
```

# Month

```python
# create two dataframes:
#   A: count occurances of each month
#   B: from Month.xlsx
# We look for the counts of the current month in 'A' dataframe and store it in a temp variable. Then we allocate the current month in 'B' dataframe and change the value of number


#           SET UP
def str_to_datetime(s):
  if isinstance(s, str):
    month_dict= {
      'Jan': 1,
      'Feb': 2,
      'Mar': 3,
      'Apr': 4,
      'May': 5,
      'Jun': 6,
      'Jul': 7,
      'Aug': 8,
      'Sep': 9,
      'Oct': 10,
      'Nov': 11,
      'Dec': 12
    }
    split = s.split(' ')
    day, month, year = 1, (month_dict[split[1]]), int('20' + split[2])
    return dt.datetime(year=year, month=month, day=day)
  else:
    return ()


#         DATAFRAME A
temp_dates = pd.date_range(start= '12/01/2016', end='12/01/2022', freq='1m').tolist()
temp_dates_= []
for d in temp_dates:
    d= d + dt.timedelta(days=1)
    d= d.date()
    temp_dates_.append(d)
temp_dates_
month_excel= pd.read_excel(input_months_filepath)
month_excel['date']= temp_dates_


#         DATAFRAME B
join_dates= pd.DataFrame(df_no_duplicate['Join Date'])
join_dates['date']= join_dates['Join Date'].apply(str_to_datetime)
monthly_join_dates= pd.DataFrame({'date': join_dates.date.value_counts().index, 'count':join_dates.date.value_counts().values})
monthly_join_dates= monthly_join_dates.drop(index=0, axis=0)
monthly_join_dates.index= np.arange(len(monthly_join_dates))


# Assignment
for i in range(66, len(month_excel)):
    temp_month_excel= month_excel.date[i]
    for j in range(len(monthly_join_dates)):
        temp_month_join= monthly_join_dates.date[j]
        if temp_month_excel.year == temp_month_join.year and temp_month_excel.month == temp_month_join.month:
            month_excel['number'][i]= monthly_join_dates['count'][j]
            print(i,'- EXCEL|',temp_month_excel.month,'/', temp_month_excel.year,': ',month_excel['number'][i])
            print(j,'- JOIN |',temp_month_join.month,'/', temp_month_join.year, ':', monthly_join_dates['count'][j])
            print('====================')
month_excel= month_excel.drop(columns=['date'])
month_excel.to_excel(input_months_filepath, index= False)
month_excel.to_excel(output_months_filepath, index= False)
```
